### Lists Method: (in built functions)
- append : add items at the end of list
- insert : add items at specific index
- removes: removes first common items
- pop    : removes & returns last item
- sort   : sorts the list in ASC order
- reverse: reverses the list
- clear  : empties the list
- index  : check the index of items in list
- count  : counts the items

```python
# Append

fruits.append("Papaya")
print(fruits)
```

```python
# Insert

fruits.insert(1,"Peach")
print(fruits)
```

```python
# Remove

fruits.remove("Peach")
print(fruits)
```

```python
# Pop

fruits.pop()
print(fruits)
```

```python
# Sort

fruits.sort()
```

```python
# Index

index=fruits.index("cherry")
print(index)
```

```python
# Count

fruits.insert(2,"banana")
print(fruits.count("banana"))
```

```python
# Clear
# Remove all items from the list

fruits.clear()
print(fruits)
```

### List comprehension
- List Comprehension is a short, fast, and readable way to create lists in Python.
- It replaces long for loops with a compact syntax.
- Basic Syntax : [expression for item in iterable]
- with conditional logic    [expression for item in iterable if condition]
- Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]

```python
##  without list comprehension
lst=[]
for x in range(10):
    lst.append(x**2)

print(lst)
```

```python
## with list comprehension

[x**2 for x in range(10)]
```

```python
### Basic List Comphrension

sqaure=[num**2 for num in range(10)]
print(sqaure)
```

```python
### List Comprehension with Condition
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)

print(lst)
```

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
## Nested List Comphrension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## List Comprehension with function calls
words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)  # Output: [5, 5, 6, 4, 13]
```

#### Conclusion
List comprehensions are a powerful and concise way to create lists in Python. They are syntactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehensions will help you write cleaner and more efficient Python code.
